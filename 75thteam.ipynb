```python
#Players with more than 1 ring and Mvp won

def extract_first_number(df, columns):
    for column in columns:
        df[column] = df[column].astype(str).str.extract('(\d+)').astype(float).fillna(0).astype(int)
    return df

# Apply the function to clean multiple columns
columns_to_clean = ['All Star', 'MVP won', 'Championships won']
df = extract_first_number(df, columns_to_clean)




Elites = df[(df["Pts"].str.replace(",", "").astype(float) > 10000) & (df["Ast"].str.replace(",","").astype(float)>3000)
        & (df["All Star"].astype(float)>5) & (df["MVP won"].astype(float)>0) & (df["Championships won"].astype(float)>0)]
# Fill NaN values
Clean_data= Elites[["All Star", "MVP won","Championships won"]].fillna(0)

# Iterate and print
for index, row in Elites.iterrows():
    print(f"{row['Name[5]']}: {row['Pts']} points; {row['Ast']} assists; {row['All Star']} All-star appearances ; {row['MVP won']} Mvp(s); {row['Championships won']} Rings")
```

```python
num_elites=len(Elites)
print(f"Elite players: {num_elites}")
```
